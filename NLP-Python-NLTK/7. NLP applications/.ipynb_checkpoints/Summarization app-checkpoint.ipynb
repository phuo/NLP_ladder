```python
f=open('../data/carl_sagan_quote.txt')
text=f.read()
text
```

```python
import nltk
```

```python
results = []
```

```python
for number, sentence in enumerate(nltk.sent_tokenize(text)):
    number_tokens=len(nltk.word_tokenize(sentence))
    tagged=nltk.pos_tag(nltk.word_tokenize(sentence))
    number_nouns=len([word for word, pos in tagged if pos in ["NN","NNP"]])
    #use NER to tag the named entities
    ners=nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sentence)),binary=False)
    number_ners=len([chunk for chunk in ners if hasattr(chunk,'node')])
    score=(number_ners+number_nouns)/float(number_tokens)
    results.append((number,number_tokens,number_nouns,score,sentence))
```

```python
print(results)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
results1=[]
```

```python
sentences1=nltk.sent_tokenize(text)
```

```python
vectorizer = TfidfVectorizer(norm='l2',min_df=0, use_idf=True, smooth_idf=False, sublinear_tf=True)
```

```python
sklearn_binary=vectorizer.fit_transform(sentences1)
```

```python
print(vectorizer.get_feature_names())
```

```python
import pandas as pd
df = pd.DataFrame(sklearn_binary.toarray().transpose(), index=vectorizer.get_feature_names())
df
```

```python
print(sklearn_binary.toarray())
```

```python
for i in sklearn_binary.toarray():
    results1.append(i.sum()/float(len(i.nonzero()[0])))
    print(results1)
```
